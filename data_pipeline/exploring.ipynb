```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("data/boardgames_enriched.csv")
```

```python
# make a single label to show non-pareto, almost-pareto, and pareto
data["label"] = "Non-Pareto"
data.loc[data["almost_pareto"], "label"] = "Almost-Pareto"
data.loc[data["is_pareto"], "label"] = "Pareto"


ax = sns.scatterplot(
    data=data, x="bayesaverage", y="complexity", hue="label", alpha=0.5
)
ax.set_title("Pareto Frontier")
ax.set_xlabel("Bayesian Average Rating")
ax.set_ylabel("Complexity")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler


scaled = MinMaxScaler().fit_transform(data[["bayesaverage", "complexity"]])

sns.scatterplot(
    x=scaled[:, 0],
    y=scaled[:, 1],
    hue=data["label"],
    alpha=0.5,
)

```
